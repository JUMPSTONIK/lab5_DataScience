```python
# Loading Data

df = pd.read_csv("../input/grammar-and-online-product-reviews/GrammarandProductReviews.csv", engine='python')
df.head()
```

```python
# Get how many manufacturers exist
df_manufacturer = df['manufacturer'].value_counts()
df_manufacturer.shape
```

```python
# Get the 20 manufacturers with the most review
df_manufacturer.head(20)
```

```python
# 6.5 and 6.6
# Take the manufacturers and rating column
df_manufacturer_rating = df[['manufacturer', 'reviews.rating']]

# List of the 20 manufacturers with the most reviews
df1 = df_manufacturer.rename_axis('manufacturer').reset_index(name='counts').head(20)
list_df1 = df1['manufacturer'].to_list()

# Take all the products that were made by the 20 manufacturers with the most reviews
top20_manufacturer_rating = df_manufacturer_rating[df_manufacturer_rating['manufacturer'].isin(list_df1)]

# Goup by manufacturer and get the average rating
df2 = top20_manufacturer_rating.groupby('manufacturer', as_index=False)['reviews.rating'].mean()
df2.sort_values(by='reviews.rating', ascending=False)
```

```python
# 6.3
df_manufacturer_username = df[['manufacturer', 'reviews.username']]
df3 = df_manufacturer_username.drop_duplicates()
df3['reviews.username'].value_counts().head(20)
```
